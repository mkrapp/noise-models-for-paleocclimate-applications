```python
%matplotlib inline 
import numpy as np
import scipy.signal
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
# https://stackoverflow.com/a/34261500/1498309
class _Lambda:

    def __init__(self, func):
        self._func = func

    def __add__(self, other):
        return _Lambda(
            lambda *args, **kwds: self._func(*args, **kwds) + other._func(*args, **kwds))

    def __call__(self, *args, **kwds):
        return self._func(*args, **kwds)
```

```python
# copied and adapted from https://gist.github.com/m-schubert/45c562146c6607b8990f1e8f34ff87b0
# (c) Matthew Schubert
class NoiseGenerator:
    """Generates gaussian noise with a specified power spectral density.
    Noise is generated by first creating a gaussian white base noise in the
    frequency domain, and then shaping it using a colouring function. Various
    colouring functions are provided for convenience.
    Examples
    --------
    >>> ng = NoiseGenerator()
    Generate 100000 white noise points sampled at 1 kHz with a PSD of 0.1.
    >>> white = ng.generate(1e-3, 100000, colour=ng.white(0.1))
    Generate some other colours of noise at a sampling rate of 1 MHz.
    >>> pink = ng.generate(1e-6, 1000000, colour=ng.pink(10.0))
    >>> blue = ng.generate(1e-6, 1000000, colour=ng.blue())
    Use a custom piecewise colouring function to generate specific noise.
    >>> frequencies = [90.0, 100.0, 110.0, 450.0, 500.0, 550.0]
    >>> psds = [0.1, 10.0, 0.01, 0.01, 2.0, 0.001]
    >>> colour = ng.piecewise_logarithmic(frequencies, psds)
    >>> custom = ng.generate(1e-4, 1000000, colour=colour)
    """

    def __init__(self,seed=None):
        self.seed = seed 
        self.rng = np.random.default_rng(self.seed)

    def generate(self, dt, n, colour=None):
        """Generates uniformly sampled noise of a particular colour.
        Parameters
        ----------
        dt : float
            Sampling period, in seconds.
        n : int
            Number of samples to generate.
        colour : function, optional
            Colouring function that specifies the PSD at a given frequency. If
            not specified, the noise returned will be white Gaussian noise with
            a PSD of 1.0 across the entire frequency range.
        Returns
        -------
        numpy.array
            Length `n` array of sampled noise.
        """

        f, x_f = self._base_noise(dt, n)

        if colour:
            x_f *= np.sqrt(colour(f))

        return np.fft.irfft(x_f)

    @staticmethod
    def white(scale=1.0):
        """Creates a white noise colouring function.
        Parameters
        ----------
        scale : float, optional
            Multiplier to adjust the scale of the white noise.
        Returns
        -------
        function
            White noise colouring function that can be used with `generate()`.
            The function returned will be :math:`y(f) = s`, where :math:`s` is
            `scale`.
        """

        return _Lambda(lambda f: scale + f*0)

    @staticmethod
    def pink(scale=1.0):
        """Creates a pink noise colouring function.
        Parameters
        ----------
        scale : float, optional
            Multiplier to adjust the scale of the pink noise.
        Returns
        -------
        function
            Pink noise colouring function that can be used with `generate()`.
            The function returned will be :math:`y(f) = s / f`, where :math:`s`
            is `scale`. At f = 0, the function will simply return 0.0.
        """

        return _Lambda(lambda f: scale / np.where(f == 0.0, np.inf, f))

    @staticmethod
    def brownian(scale=1.0):
        """Creates a brownian noise colouring function.
        Parameters
        ----------
        scale : float, optional
            Multiplier to adjust the scale of the brownian noise.
        Returns
        -------
        function
            Brownian noise colouring function that can be used with
            `generate()`. The function returned will be
            :math:`y(f) = s / f ^ 2`, where :math:`s` is `scale`. At f = 0, the
            function will simply return 0.0.
        """

        return _Lambda(lambda f: scale / np.where(f == 0.0, np.inf, f**2))

    @staticmethod
    def blue(scale=1.0):
        """Creates a blue noise colouring function.
        Parameters
        ----------
        scale : float, optional
            Multiplier to adjust the scale of the blue noise.
        Returns
        -------
        function
            Blue noise colouring function that can be used with `generate()`.
            The function returned will be :math:`y(f) = s * f`, where :math:`s`
            is `scale`.
        """

        return _Lambda(lambda f: scale * f)

    @staticmethod
    def violet(scale=1.0):
        """Creates a violet noise colouring function.
        Parameters
        ----------
        scale : float, optional
            Multiplier to adjust the scale of the violet noise.
        Returns
        -------
        function
            Violet noise colouring function that can be used with `generate()`.
            The function returned will be :math:`y(f) = s * f ^ 2`, where
            :math:`s` is `scale`.
        """

        return _Lambda(lambda f: scale * f**2)
    
    @staticmethod
    def powerlaw(scale=1.0, x=1.0):
        """Creates a power-law noise colouring function.
        Parameters
        ----------
        scale : float, optional
            Multiplier to adjust the scale of the noise.
        x : float, optional
            Exponent to adjust the scale of the noise.

        Returns
        -------
        function
            Power-law noise colouring function that can be used with
            `generate()`. The function returned will be
            :math:`y(f) = s / f ^ x`, where :math:`s` is `scale`. At f = 0, the
            function will simply return 0.0.
        """

        return _Lambda(lambda f: scale / np.where(f == 0.0, np.inf, f**x))

    # Some common aliases for the various colours of noise
    brown = brownian
    red = brownian
    azure = blue
    purple = violet

    @staticmethod
    def piecewise_logarithmic(frequencies, psds):
        """Creates a custom piecewise colouring function
        Parameters
        ----------
        frequencies : numpy.array
            Array of frequencies, in Hz
        psds : numpy.array
            Array of PSDs
        Returns
        -------
        function
            Custom noise colouring function that can be used with `generate()`.
            The function is linearly interpolated in log space for the given
            frequencies and PSDs. Values outside the range of frequencies given
            are set to the PSD of the closest endpoint.
        """

        # Convert to log space
        log_frequencies = np.log(frequencies)
        log_psds = np.log(psds)

        # Create a closure for our colour function that suppresses the warning
        # about np.log(0) (which correctly returns -np.inf anyway)
        def colour(f):
            with np.errstate(divide="ignore"):
                return np.exp(np.interp(np.log(f), log_frequencies, log_psds))

        return _Lambda(colour)

    def _base_noise(self, dt, n):
        """Produces a frequency domain representation of uniformly sampled
        Gaussian white noise.
        """

        # Calculate random frequency components
        f = np.fft.rfftfreq(n, dt)
        x_f = self.rng.normal(0,
            0.5, len(f)) + 1j * self.rng.normal(0, 0.5, len(f))
        x_f *= np.sqrt(n / dt)

        # Ensure our 0 Hz and Nyquist components are purely real
        x_f[0] = np.abs(x_f[0])
        if len(f) % 2 == 0:
            x_f[-1] = np.abs(x_f[-1])

        return f, x_f
```

```python
def plot_psd(x_t, dt, **kwargs):
    """Plots a signal and its PSD given samples and sampling period."""

    t = np.linspace(0.0, dt * len(x_t), len(x_t), endpoint = False)
    f, s_f = scipy.signal.welch(x_t, fs=1/dt, **kwargs)

    fig, ax = plt.subplots(2, 1)
    ax[0].plot(t, x_t, linewidth=0.5)
    ax[0].set_xlabel("Time (a)")
    ax[0].set_ylabel("Value (V)")
    ax[1].loglog(f, s_f, linewidth=0.5)
    ax[1].set_xlabel("Frequency (1/a)")
    ax[1].set_ylabel("PSD ($V^2/(1/a)$)")
    fig.tight_layout()
    return fig, ax
```

```python
dt = 2500.0
n = 2000
#nperseg = 512
nperseg = 2**np.log2(n/4).astype(int)
ng = NoiseGenerator(1234)
print(n,dt,nperseg)
```

```python
white = ng.generate(dt, n, colour=ng.white(0.1))
fig, ax = plot_psd(white, dt, nperseg=nperseg)
x = ax[1].lines[0].get_xdata()
l, = ax[1].plot(x,ng.white(0.1)(x))
```

```python
pink = ng.generate(dt, n, colour=ng.pink())
fig, ax = plot_psd(pink, dt, nperseg=nperseg)
x = ax[1].lines[0].get_xdata()
l, = ax[1].plot(x,ng.pink()(x))
```

```python
blue = ng.generate(dt, n, colour=ng.blue())
fig, ax = plot_psd(blue, dt, nperseg=nperseg)
x = ax[1].lines[0].get_xdata()
l, = ax[1].plot(x,ng.blue()(x))
```

```python
violet = ng.generate(dt, n, colour=ng.violet())
fig, ax = plot_psd(violet, dt, nperseg=nperseg)
x = ax[1].lines[0].get_xdata()
l, = ax[1].plot(x,ng.violet()(x))
```

```python
brownian = ng.generate(dt, n, colour=ng.brownian())
fig, ax = plot_psd(brownian, dt, nperseg=nperseg)
x = ax[1].lines[0].get_xdata()
l, = ax[1].plot(x,ng.brownian()(x))
```

```python
powerlaw = ng.generate(dt, n, colour=ng.powerlaw(x=1.5))
fig, ax = plot_psd(powerlaw, dt, nperseg=nperseg)
x = ax[1].lines[0].get_xdata()
l, = ax[1].plot(x,ng.powerlaw(x=1.5)(x))
```

```python
# mixed
periodic = _Lambda(lambda f: np.where(np.logical_and(f<1/100,f>1/90),10,f))
mixed = ng.generate(dt, n, colour=ng.blue())
fig, ax = plot_psd(mixed, dt, nperseg=nperseg)
x = ax[1].lines[0].get_xdata()
l, = ax[1].plot(x,(ng.blue())(x))
```

```python
# mixed
periodic1 = _Lambda(lambda f: np.where(np.logical_and(f>1/(100000+3*dt),f<1/(100000-3*dt)),100,1))
periodic2 = _Lambda(lambda f: np.where(np.logical_and(f>1/(41000+dt),f<1/(41000-dt)),50,1))
periodic3 = _Lambda(lambda f: np.where(np.logical_and(f>1/(26000+0.5*dt),f<1/(26000-0.5*dt)),25,1))
periodic = periodic1+periodic2+periodic3
mixed = ng.generate(dt, n, colour=periodic+ng.pink(scale=1e-4))
fig, ax = plot_psd(mixed, dt, nperseg=nperseg)
x = ax[1].lines[0].get_xdata()
l, = ax[1].plot(x,(periodic+ng.pink(scale=1e-4))(x))
```

```python
# Reproducing Figures 1 and 2 from
# Vaughan, S., R. J. Bailey, and D. G. Smith. 2011.
# Detecting Cycles in Stratigraphic Data: Spectral Analysis in the Presence of Red Noise. 
# Paleoceanography 26 (4). https://doi.org/10.1029/2011PA002195.

dt = 1.0
n = 600
nperseg = 256
ng = NoiseGenerator(892022)
print(n,dt,nperseg)

f0 = 0.015
df = f0/100
print(df)

noises = {
    "Periodic": _Lambda(lambda f: np.where(np.logical_and(f>=f0-df,f<=f0+df),5,1e-4)),
    "White noise": ng.white(0.04),
    "Red noise": ng.powerlaw(scale=1e-2,x=1)
}

noises["Mixed"] = noises["Periodic"] + noises["White noise"] + noises["Red noise"]

fig1, ax1 = plt.subplots(2, 2,sharex=True,sharey=True, figsize=(8,4))
fig2, ax2 = plt.subplots(2, 2,sharex=True,sharey=True, figsize=(8,4))
ax1 = ax1.flatten()
ax2 = ax2.flatten()
for i,(name, noise) in enumerate(noises.items()):
#     for _ in range(100):
#         x_t = ng.generate(dt, n, colour=noise)
#         t = np.linspace(0.0, dt * len(x_t), len(x_t), endpoint = False)
#         ax1[i].plot(t, x_t, 'r-', linewidth=0.1)
#         f, s_f = scipy.signal.welch(x_t, fs=1/dt, nperseg=nperseg)
#         ax2[i].loglog(f, s_f, 'r-', linewidth=0.1)
    x_t = ng.generate(dt, n, colour=noise)
    t = np.linspace(0.0, dt * len(x_t), len(x_t), endpoint = False)
    ax1[i].plot(t, x_t, lw=1)
    f, s_f = scipy.signal.welch(x_t, fs=1/dt, nperseg=nperseg)
    ax2[i].loglog(f, s_f, lw=1)
    f = np.logspace(-3.1,1.1,1000)
    ax2[i].plot(f,noise(f),'k-',lw=2)
    ax1[i].set_title(name,fontsize=10)
    ax2[i].set_title(name,fontsize=10)
    ax1[i].set_ylim(-1.1,1.1)
    ax2[i].set_ylim(0.009,11)
    ax2[i].set_xlim(0.9e-3,1.1)
    
ax2[-1].plot(f,noises['Red noise'](f),'k--',lw=1,zorder=0)
ax2[-1].plot(f,noises['White noise'](f),'k--',lw=1,zorder=0)

fig1.tight_layout()
fig2.tight_layout()
fig1.supylabel('x(t)',x=0.0)
fig1.supxlabel('Time', y=0.0)
fig2.supylabel('Power density',x=0.0)
fig2.supxlabel('Frequency',y=0.0);
```
